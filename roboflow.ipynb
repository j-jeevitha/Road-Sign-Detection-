```python
#  Integrate with OpenCV for AR Display
```

```python
import cv2
from ultralytics import YOLO

# Load trained YOLO model
model = YOLO()

# Open webcam
cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    # Run YOLO detection
    results = model(frame)

    # Draw bounding boxes and labels
    for r in results:
        for box in r.boxes:
            x1, y1, x2, y2 = box.xyxy[0]  # Bounding box coordinates
            label = model.names[int(box.cls[0])]  # Class label
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
            cv2.putText(frame, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
    
    cv2.imshow("Road Sign Detection", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# Add Augmented Reality (AR) Features
# Basic AR Overlay with OpenCV
```

```python
import cv2
import numpy as np
from ultralytics import YOLO

# Load the trained YOLO model
model = YOLO("C:\\Users\\chitr\\OneDrive\\Documents\\Desktop\\JEEVITHA\\finalproject\\runs\\detect\\train22\\weights\\best.pt")

# Open webcam
cap = cv2.VideoCapture(0)

# Load AR road sign overlay (Ensure it's a PNG with transparency)
ar_overlay = cv2.imread("C:\\Users\\chitr\\OneDrive\\Documents\\Desktop\\JEEVITHA\\finalproject\\train\images\\IMG-20250216-WA0047_jpg.rf.83a4b709eb7147d464a9fdaeaeb834b7.jpg", cv2.IMREAD_UNCHANGED)  

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Perform YOLO detection
    results = model(frame)

    for r in results:
        for box in r.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])  # Bounding box
            label = model.names[int(box.cls[0])]  # Detected sign label

            # Resize AR overlay to match detected sign size
            overlay_resized = cv2.resize(ar_overlay, (x2 - x1, y2 - y1))

            # Blend overlay with the frame using alpha blending
            alpha = 0.6  # Adjust transparency
            frame[y1:y2, x1:x2] = cv2.addWeighted(frame[y1:y2, x1:x2], 1 - alpha, overlay_resized, alpha, 0)

            # Display label
            cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)

    # Show AR detection result
    cv2.imshow("AR Road Sign Detection", frame)

    # Quit if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import pygame
from OpenGL.GL import *
from OpenGL.GLUT import *

def draw_text(position, text, font_size=18, color=(1, 1, 1)):
    """Draws AR text on the HUD"""
    glColor3f(*color)
    glRasterPos2f(*position)
    for char in text:
        glutBitmapCharacter(GLUT_BITMAP_HELVETICA_18, ord(char))

def draw_hud():
    """Draws a simple AR HUD overlay"""
    glBegin(GL_LINES)
    glVertex2f(-0.5, 0.8)
    glVertex2f(0.5, 0.8)
    glEnd()
    draw_text((-0.1, 0.85), "Road Sign Ahead!")

pygame.init()
display = (800, 600)
pygame.display.set_mode(display, pygame.DOUBLEBUF | pygame.OPENGL)

glutInit()
glOrtho(-1, 1, -1, 1, -1, 1)

running = True
while running:
    glClear(GL_COLOR_BUFFER_BIT | GL_DEPTH_BUFFER_BIT)
    draw_hud()
    pygame.display.flip()

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

pygame.quit()
```

```python
import pygame
from OpenGL.GL import *
from OpenGL.GLUT import *
from OpenGL.GLU import *

def draw_text(position, text, color=(1, 1, 1)):
    """Draws text on the AR HUD"""
    glColor3f(*color)
    glRasterPos3f(*position)
    for ch in text:
        glutBitmapCharacter(GLUT_BITMAP_HELVETICA_18, ord(ch))

def display():
    """HUD Display"""
    glClear(GL_COLOR_BUFFER_BIT | GL_DEPTH_BUFFER_BIT)
    glLoadIdentity()

    draw_text((-0.5, 0.5, 0), "SPEED LIMIT: 40 km/h", color=(1, 0, 0))
    draw_text((-0.5, 0.3, 0), "STOP SIGN AHEAD!", color=(1, 1, 0))

    pygame.display.flip()

# Initialize Pygame & OpenGL
pygame.init()
display = (800, 600)
pygame.display.set_mode(display, pygame.DOUBLEBUF | pygame.OPENGL)
gluPerspective(45, (display[0] / display[1]), 0.1, 50.0)
glTranslatef(0.0, 0.0, -5)

# Run OpenGL Loop
while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            quit()
    
    display()
```
